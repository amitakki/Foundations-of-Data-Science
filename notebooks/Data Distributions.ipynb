### Each attribute or variable within a dataset has its own distribution, describing how the values of that attribute are spread or distributed. The type of distribution can vary based on the nature of the data. Here are a few common types of distributions:

### Normal Distribution (Gaussian Distribution):

* Symmetrical and bell-shaped.
* Characterized by mean and standard deviation.
* Many natural phenomena follow a normal distribution.
* **Probability Density Function (PDF):** \
$ f(x | \mu, \sigma^2) = \frac{1}{\sqrt{2\pi\sigma^2}} \exp\left(-\frac{(x-\mu)^2}{2\sigma^2}\right) $

* **Cumulative Distribution Function (CDF):** \
$ F(x | \mu, \sigma^2) = \frac{1}{2}\left[1 + \text{erf}\left(\frac{x-\mu}{\sigma\sqrt{2}}\right)\right]$

### Uniform Distribution:

* All values have the same probability.
* Often seen in random or uniformly generated data.
* **Probability Density Function (PDF):**
$ f(x | a, b) = \frac{1}{b - a} \ for\  a \leq x \leq b,\ and
\\ f(x | a, b) = 0\ otherwise.$

* **Cumulative Distribution Function (CDF):**
$ F(x | a, b) = \frac{x - a}{b - a}\ for\ a \leq x \leq b, \\
F(x | a, b) = 0\ for\ x < a, and\\
F(x | a, b) = 1\ for\ x > b.$

### Exponential Distribution:

* Describes the time between events in a Poisson process.
* Often used in survival analysis.
* **Probability Density Function (PDF):** \
$ f(x | \lambda) = \lambda e^{-\lambda x} \ for \ x \geq 0\, and \\ f(x | \lambda) = 0\ for \ x < 0. $

* **Cumulative Distribution Function (CDF):** \
$ F(x | \lambda) = 1 - e^{-\lambda x} \ for \ x \geq 0\, and\\  F(x | \lambda) = 0\ for \ x < 0. $


### Log-Normal Distribution:

* Results when the logarithm of a variable has a normal distribution.
* Common in financial data and certain biological measurements.
* **Probability Density Function (PDF):**
$ f(x | \mu, \sigma) = \frac{1}{x\sigma\ \sqrt{2\pi}} \exp\left(-\frac{(\ln x - \mu)^2}{2\sigma^2}\right) \ for \ x > 0\ ,\ and \\ f(x | \mu, \sigma) = 0\ for\ x \leq 0. $

* **Cumulative Distribution Function (CDF):**
No simple closed-form expression.

### Poisson Distribution:

* Describes the number of events occurring in fixed intervals.
* Often used for count data.
* **Probability Mass Function (PMF):** \
$ P(X=k | \lambda) = \frac{e^{-\lambda}\lambda^k}{k!} \ for \ k=0,1,2,\ldots $

* **Cumulative Distribution Function (CDF):** \
$ F(k | \lambda) = \sum_{i=0}^k \frac{e^{-\lambda}\lambda^i}{i!} \ for \ k=0,1,2,\ldots $


### Binomial Distribution:

* Describes the number of successes in a fixed number of trials.
* Common in binary outcome scenarios.
* **Probability Mass Function (PMF):** \
$ P(X=k | n, p) = \binom{n}{k} p^k (1-p)^{n-k} \ for \ k=0,1,2,\ldots,n. $

* **Cumulative Distribution Function (CDF):** \
$ F(k | n, p) = \sum_{i=0}^k \binom{n}{i} p^i (1-p)^{n-i} \ for \ k=0,1,2,\ldots,n. $

### Gamma Distribution:

* Generalization of the exponential distribution.
* Used in various fields, including queuing theory and physics.
* **Probability Density Function (PDF):** \
$ f(x | \alpha, \beta) = \frac{\beta^\alpha x^{\alpha-1} e^{-\beta x}}{\Gamma(\alpha)} \ for \ x \geq 0\ , and \\
 (f(x | \alpha, \beta) = 0\ for \ x < 0\ . $

* **Cumulative Distribution Function (CDF):** \
$ F(x | \alpha, \beta) = \frac{1}{\Gamma(\alpha)} \gamma(\alpha, \beta x) \ , where \ \gamma(\alpha, \beta x)\ is\ the\ lower\ incomplete\ gamma\ function.$


### Beta Distribution:

* Describes random variables that are constrained to the interval [0, 1].
* Commonly used in Bayesian statistics.
* **Probability Density Function (PDF):** \
$ f(x | \alpha, \beta) = \frac{x^{\alpha-1}(1-x)^{\beta-1}}{B(\alpha, \beta)} \ for \ 0 \leq x \leq 1\ , and \\
  f(x | \alpha, \beta) = 0,\ otherwise. $

* **Cumulative Distribution Function (CDF):** \
$ F(x | \alpha, \beta) = I_x(\alpha, \beta) \, where \ I_x(\alpha, \beta)\ is\ the\ regularized\ incomplete\ beta\ function. $


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Sample size
sample_size = 10000
```

```python
# Normal Distribution (Gaussian)
mu, sigma = 0, 1
normal_samples = np.random.normal(mu, sigma, sample_size)
```

```python
# Uniform Distribution
uniform_samples = np.random.uniform(0, 1, sample_size)
```

```python
# Exponential Distribution
lambda_param = 0.5
exponential_samples = np.random.exponential(1/lambda_param, sample_size)
```

```python
# Log-Normal Distribution
lognormal_samples = np.random.lognormal(mu, sigma, sample_size)
```

```python
# Poisson Distribution
poisson_samples = np.random.poisson(5, sample_size)
```

```python
# Binomial Distribution
n, p = 10, 0.5
binomial_samples = np.random.binomial(n, p, sample_size)
```

```python
# Gamma Distribution
alpha, beta = 2, 2
gamma_samples = np.random.gamma(alpha, beta, sample_size)
```

```python
# Beta Distribution
alpha, beta = 2, 5
beta_samples = np.random.beta(alpha, beta, sample_size)
```

```python
# Plotting histograms for visualization
plt.figure(figsize=(12, 8))

plt.subplot(3, 3, 1)
plt.hist(normal_samples, bins=30, color='blue', alpha=0.7)
plt.title('Normal Distribution')

plt.subplot(3, 3, 2)
plt.hist(uniform_samples, bins=30, color='orange', alpha=0.7)
plt.title('Uniform Distribution')

plt.subplot(3, 3, 3)
plt.hist(exponential_samples, bins=30, color='green', alpha=0.7)
plt.title('Exponential Distribution')

plt.subplot(3, 3, 4)
plt.hist(lognormal_samples, bins=30, color='red', alpha=0.7)
plt.title('Log-Normal Distribution')

plt.subplot(3, 3, 5)
plt.hist(poisson_samples, bins=30, color='purple', alpha=0.7)
plt.title('Poisson Distribution')

plt.subplot(3, 3, 6)
plt.hist(binomial_samples, bins=30, color='brown', alpha=0.7)
plt.title('Binomial Distribution')

plt.subplot(3, 3, 7)
plt.hist(gamma_samples, bins=30, color='pink', alpha=0.7)
plt.title('Gamma Distribution')

plt.subplot(3, 3, 8)
plt.hist(beta_samples, bins=30, color='gray', alpha=0.7)
plt.title('Beta Distribution')

plt.tight_layout()
plt.show()
```
